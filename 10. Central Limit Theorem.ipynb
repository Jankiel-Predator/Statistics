# Central Limit Theorem

### Objective
The aim of this experiment is to investigate and verify the Central Limit Theorem using random numbers generator, using uniform and exponential distribution samples.

## 1. Theorem content
The **Central Limit Theorem** states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the original distribution.

## 2. Conclusions
In both cases, this Theorem is valid and clearly the respective distributions added converge to normal distribution, as proved by the probplot. The only difference is that in the case of uniform distribution, this convergence is faster than for the exponential distribution. That means, adding 10 of the subsequent distributions in the case of uniform distribution was sufficient to obtain visually appealing normal histogram. For the exponential distribution it performed a little worse, as illustrated by the normal probability plot. We can see that adding uniform distribuions only 3 times was already enough to resemble the normal curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import probplot

plt.style.use("fivethirtyeight")
np.random.seed(24)
```

```python
### Number of bins to plot histograms

num_data_points = 500
num_bins = int(np.sqrt(num_data_points))
```

# Uniform Distribution

```python
### Sample of U(0, 1) distribution

F1 = np.random.uniform(size=num_data_points)
```

```python
plt.figure(figsize=(14, 7))
plt.hist(F1, bins=num_bins, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histogram for the Uniform Distribution of the Initial Sample (F1)')
plt.ylabel('Probability Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
### Add subsequent data points to see whether they start to resemble normal distribution

fig, axes = plt.subplots(3, 3, figsize=(16, 12))
fig.suptitle('Histograms for the Sum of 2 to 10 Samples')
for i in range(2, 11):
    samples = [np.random.uniform(size=num_data_points) for _ in range(i)]
    F_sum = np.sum(samples, axis=0)

    ax = axes[(i-2)//3, (i-2)%3]
    ax.hist(F_sum, bins=num_bins, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    ax.set_title(f'Uniform Distribution of F{i}')
    ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Verify whether the final distribution is normal

plt.figure(figsize=(14,7))
probplot_data_F10 = probplot(F_sum, plot=plt)
plt.title('Normal Probability Plot for the Cumulated Sum of All Previous Distributions')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.show()
```

```python
mean_F1 = np.mean(F1)
std_dev_F1 = np.std(F1)
mean_F10 = np.mean(F_sum)
std_dev_F10 = np.std(F_sum)

print(f"Mean of sum of initial U(0,1): {mean_F1:.4f}")
print(f"Mean of sum of distributions: {mean_F10:.4f}\n")
print(f"Standard Deviation of initial U(0,1): {std_dev_F1:.4f}")
print(f"Standard Deviation of sum of distributions: {std_dev_F10:.4f}")
```

# Exponential Distribution

```python
### Sample of Exp(1)

lambda_val = 1
F1_exp = np.random.exponential(scale=lambda_val, size=num_data_points)
```

```python
plt.figure(figsize=(14, 7))
plt.hist(F1_exp, bins=num_bins, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histogram for the Exponential Distribution of the Initial Sample (F1)')
plt.ylabel('Probability Density')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
### Add subsequent data points to see whether they start to resemble normal distribution

fig, axes_exp = plt.subplots(3, 3, figsize=(16, 12))
fig.suptitle('Histograms for the Sum of 2 to 10 Samples')

for i in range(2, 11):
    samples_exp = [np.random.exponential(scale=1/lambda_val, size=num_data_points) for _ in range(i)]
    F_sum_exp = np.sum(samples_exp, axis=0)

    ax = axes_exp[(i-2)//3, (i-2)%3]
    ax.hist(F_sum_exp, bins=num_bins, density=True, alpha=0.7, color='coral', edgecolor='black')
    ax.set_title(f'Exponential Distribution of F{i}')
    ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Verify whether the final distribution is normal

plt.figure(figsize=(14,7))
probplot_data_F10 = probplot(F_sum_exp, plot=plt)
plt.title('Normal Probability Plot for the Cumulated Sum of All Previous Distributions')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.show()
```

```python
mean_F1_exp = np.mean(F1_exp)
std_dev_F1_exp = np.std(F1_exp)
mean_F10_exp = np.mean(F_sum_exp)
std_dev_F10_exp = np.std(F_sum_exp)

print(f"Mean of sum of initial Exp(1): {mean_F1_exp:.4f}")
print(f"Mean of sum of distributions: {mean_F10_exp:.4f}\n")
print(f"Standard Deviation of initial Exp(1): {std_dev_F1_exp:.4f}")
print(f"Standard Deviation of sum of distributions: {std_dev_F10_exp:.4f}")
```

In both cases, this Theorem is valid and clearly the respective distributions added converge to normal distribution, as proved by the probplot. The only difference is that in the case of uniform distribution, this convergence is faster than for the exponential distribution. That means, adding 10 of the subsequent distributions in the case of uniform distribution was sufficient to obtain visually appealing normal histogram. For the exponential distribution it performed a little worse, as illustrated by the normal probability plot. We can see that adding uniform distribuions only 3 times was already enough to resemble the normal curve.
