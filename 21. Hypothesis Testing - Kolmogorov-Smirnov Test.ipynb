# Hypothesis Testing - Kolmogorov-Smirnov Test

### Objective
The objective is to introduce the application of the Kolmogorov-Smirnov test for one-sample and two-sample cases, aiming to enhance the understanding of statistical hypothesis testing.

## 1. Kolmogorov-Smirnov Test
The **Kolmogorov-Smirnov test** is used to test the hypothesis that observed data follows a specified theoretical distribution. This test compares the empirical cumulative distribution function (ECDF) of the observed data with the theoretical cumulative distribution function (CDF). We consider  Kolmogorov-Smirnov tests for verifying that a sample comes from a population with some known distribution and also that two populations have the same distribution.

## 2. Formulation of Hypotheses

### One-Sample Test
**Null Hypothesis (H₀):** The empirical distribution of the sample follows the specified theoretical distribution.

**Alternative Hypothesis (H₁):** The empirical distribution of the sample does not follow the specified theoretical distribution.

### Two-Sample Test
**Null Hypothesis (H₀):** The empirical distributions of the two samples follow the same theoretical distribution.

**Alternative Hypothesis (H₁):** The empirical distributions of the two samples do not follow the same theoretical distribution.

## 3. Empirical and Theoretical Distribution Function of One Sample

### Sample 1
#### Standard Uniform Distribution:

$$ U(0, 1) := 
\begin{cases} 
1 & \text{if } 0 \leq x \leq 1 \\
0 & \text{otherwise}
\end{cases}
$$

**Null Hypothesis (H₀):** The observed data follows the $U(0,1)$ distribution.

**Alternative Hypothesis (H₁):** The observed data does not follow $U(0,1)$ distribution.

KS Test for Sample 1 vs $U(0,1)$:
KS Statistic: 0.57, p-value: 2.9887340529623177e-31
Decision: Reject H₀

#### Exponential Distribution:

$$ Exp(1) := 
\begin{cases} 
e^{-x} & \text{if } x \geq 0 \\
0 & \text{if } x < 0
\end{cases}
$$

**Null Hypothesis (H₀):** The observed data follows the $Exp(1)$ distribution.

**Alternative Hypothesis (H₁):** The observed data does not follow the $Exp(1)$ distribution.

KS Test for Sample 1 vs $Exp(1)$:
KS Statistic: 0.85, p-value: 1.429059699872966e-82
Decision: Reject H₀

#### Standard Normal Distribution:

$$ N(0, 1) := \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} $$

**Null Hypothesis (H₀):** The observed data follows the $N(0,1)$ distribution.

**Alternative Hypothesis (H₁):** The observed data does not follow the $N(0,1)$ distribution.

KS Test for Sample 1 vs $N(0,1)$:
KS Statistic: 0.09277701334036081, p-value: 0.33475021955244866
Decision: Fail to Reject H₀

## 4. Empirical Distribution Functions of Two Samples

### Sample 1 vs Sample 2
**Null Hypothesis (H₀):** The two samples follow the same theoretical distribution.

**Alternative Hypothesis (H₁):** The two samples do not follow the same theoretical distribution.

KS Test for Sample 1 and Sample 2:
KS Statistic: 0.14, p-value: 0.1211924131072925
Decision: Fail to Reject H₀

## 5. Brief Conclusions

- For Sample 1, the Kolmogorov-Smirnov test rejects the null hypothesis when comparing against both the $U(0,1)$ and $Exp(1)$ distributions, indicating that Sample 1 does not follow these distributions. However, the test fails to reject the null hypothesis when comparing against the $N(0,1)$ distribution.
- For Sample 2, the Kolmogorov-Smirnov test fails to reject the null hypothesis when comparing against the $U(0,1)$ distribution, suggesting that Sample 2 follows this distribution.
- When comparing the empirical distribution functions of Sample 1 and Sample 2, the test rejects the null hypothesis, indicating that the two samples likely do not follow the same theoretical distribution.
