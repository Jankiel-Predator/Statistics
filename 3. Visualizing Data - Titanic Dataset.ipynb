```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")
```

```python
### Load the Titanic dataset

titanic = sns.load_dataset("titanic")
```

```python
### Colors used for visualizations

colors = ["skyblue", "salmon", "lightgreen"]
```

# Percentage of passengers by class

```python
### Percentage of Passengers by Class (entire population)

population_size = len(titanic)
percentage_by_class = pd.DataFrame(titanic['class'].value_counts() / population_size * 100)
percentage_by_class.columns = ['Percentage']
percentage_by_class['Class'] = percentage_by_class.index

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Percentage of Passengers by Class (Entire Population)")

# Bar chart for the entire population
ax1.bar(percentage_by_class['Class'], percentage_by_class['Percentage'], color=colors)
ax1.set_xlabel("Class")
ax1.set_ylabel("Percentage")

# Pie chart for the entire population
ax2.pie(percentage_by_class['Percentage'], labels=percentage_by_class['Class'], autopct='%1.1f%%', startangle=90, colors=colors)
ax2.axis('equal')

plt.tight_layout()
plt.show()
```

### Percentage of Passengers by Class (Entire Population)

1. **Bar Chart**:
   - The bar chart shows that the majority of passengers traveled in Third Class, accounting for over half of the population at 55.1%.
   - First Class passengers make up a smaller portion at 24.2%.
   - Second Class has the smallest proportion of passengers, with 20.7% of the total.

2. **Pie Chart**:
   - The pie chart visually confirms the distribution seen in the bar chart.
   - Third Class is the largest segment, occupying more than half of the pie.
   - First Class and Second Class are smaller segments, with First Class slightly larger than Second Class.

```python
### Percentage of Passengers by Class (titanic of size 50)

sample_size = 50
sample = titanic.sample(sample_size, random_state=42)

sample_percentage_by_class = pd.DataFrame(sample['class'].value_counts() / sample_size * 100)
sample_percentage_by_class.columns = ['Percentage']
sample_percentage_by_class['Class'] = sample_percentage_by_class.index

fig, (ax3, ax4) = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Percentage of Passengers by Class (Sample of Size 50)")

# Bar chart for the sample
ax3.bar(sample_percentage_by_class['Class'], sample_percentage_by_class['Percentage'], color=colors)
ax3.set_xlabel("Class")
ax3.set_ylabel("Percentage")

# Pie chart for the sample
ax4.pie(sample_percentage_by_class['Percentage'], labels=sample_percentage_by_class['Class'],
        autopct='%1.1f%%', startangle=90, colors=colors)
ax4.axis('equal')

plt.tight_layout()
plt.show()
```

### Percentage of Passengers by Class (Sample of Size 50)

1. **Bar Chart**:
   - The bar chart for the sample indicates that a significant majority of passengers are in Third Class, comprising 64% of the sample.
   - First Class passengers make up 24% of the sample.
   - Second Class passengers account for 12%, the smallest proportion in the sample.

2. **Pie Chart**:
   - The pie chart for the sample mirrors the distribution shown in the bar chart.
   - Third Class is the dominant segment, reflecting a larger portion compared to the entire population.
   - First Class and Second Class make up smaller segments, with First Class twice as large as Second Class.

# Age distribution

```python
### Distribution of age of passengers (sample of size 50)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharex=True)
fig.suptitle("Age Distribution (Entire Population)", fontsize=22)

# Stem-And-Leaf Plot
markerline, stemlines, baseline = ax1.stem(titanic["age"].value_counts().index, titanic["age"].value_counts().values, 
                                           markerfmt='D', linefmt='b-', basefmt='r-')
ax1.set_title("Stem-And-Leaf Plot", fontsize=18)
ax1.set_xlabel("Age")
ax1.set_ylabel("Count")

# Histograms
ax2.hist(titanic[titanic["sex"] == "male"]["age"], alpha=0.5, label="Male", bins=10, color='blue')
ax2.hist(titanic[titanic["sex"] == "female"]["age"], alpha=0.5, label="Female", bins=10, color='red')
ax2.set_title("Histogram By Gender", fontsize=18)
ax2.legend()
ax2.set_xlabel("Age")

plt.tight_layout()
plt.show()
```

### Age Distribution (Entire Population)
1. **Stem-and-Leaf Plot**:
   - The age distribution shows a high frequency of younger passengers, especially those in their 20s and 30s.
   - There are fewer passengers in the older age groups, with a noticeable decline in counts as age increases beyond 40.
   - A few outliers exist in the age groups above 60.

2. **Histogram by Gender**:
   - The histogram reveals that the majority of passengers were male.
   - The peak age for males is around 20-30 years, while for females, it is slightly lower, around 20-25 years.
   - The distribution shows a similar pattern of more younger passengers and fewer older passengers for both genders.

```python
### Distribution of age of passengers (sample of size 50)

male_sample = sample[sample["sex"] == "male"]
female_sample = sample[sample["sex"] == "female"]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), sharex=True, sharey=True)
fig.suptitle("Age Distribution (Sample of Size 50)", fontsize=22)

# Stem-And-Leaf Plot
markerline, stemlines, baseline = ax1.stem(sample["age"].value_counts().index, sample["age"].value_counts().values, 
                                           markerfmt='D', linefmt='b-', basefmt='r-')
plt.setp(baseline, color='red', linewidth=10)
ax1.set_title("Stem-And-Leaf Plot", fontsize=18)
ax1.set_xlabel("Age")
ax1.set_ylabel("Count")

# Histograms
ax2.hist(male_sample["age"], alpha=0.5, label="Male", bins=10, color='blue')
ax2.hist(female_sample["age"], alpha=0.5, label="Female", bins=10, color='red')
ax2.set_title("Histogram By Gender", fontsize=18)
ax2.legend()
ax2.set_xlabel("Age")

plt.tight_layout()
plt.show()
```

### Age Distribution (Sample of Size 50)
1. **Stem-and-Leaf Plot**:
   - The sample shows more variability due to the smaller size, but still indicates a concentration of passengers in their 20s and 30s.
   - There are fewer data points, which means the plot is less smooth compared to the entire population's plot.

2. **Histogram by Gender**:
   - The sample histogram also shows more males than females, though the counts are much lower.
   - The age distribution for males peaks around 20-30 years, consistent with the entire population.
   - The sample includes passengers up to age 50, but the higher age groups have very few representatives, making it less representative of the older passengers seen in the entire population.

# Fare distribution

```python
### Histogram comparing fare for men, women and children and histogram comparing fare with classes (entire population)

fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharex=True)
plt.suptitle("Fare distribution (Entire Population)", fontsize=22)

# Fare distribution for men, women, and children
plt.subplot(1, 2, 1)
plt.title("Men, Women, and Children", fontsize=18)
plt.xlabel("Fare")
plt.ylabel("Frequency")
sns.histplot(data=titanic[titanic['who'] == 'man'], x='fare', bins=int(np.sqrt(len(titanic))), 
             kde=True, color=colors[0], label='Men')
sns.histplot(data=titanic[titanic['who'] == 'woman'], x='fare', bins=int(np.sqrt(len(titanic))), 
             kde=True, color=colors[1], label='Women')
sns.histplot(data=titanic[titanic['who'] == 'child'], x='fare', bins=int(np.sqrt(len(titanic))), 
             kde=True, color=colors[2], label='Children')
plt.legend()

# Fare distribution for different classes
plt.subplot(1, 2, 2)
plt.title("By Class", fontsize=18)
plt.xlabel("Fare")
plt.ylabel("Frequency")
sns.histplot(data=titanic[titanic['class'] == 'First'], x='fare', bins=int(np.sqrt(len(titanic))), 
             kde=True, color=colors[0], label='First Class')
sns.histplot(data=titanic[titanic['class'] == 'Second'], x='fare', bins=int(np.sqrt(len(titanic))), 
             kde=True, color=colors[1], label='Second Class')
sns.histplot(data=titanic[titanic['class'] == 'Third'], x='fare', bins=int(np.sqrt(len(titanic))), 
             kde=True, color=colors[2], label='Third Class')
plt.legend()

plt.tight_layout()
plt.show()
```

### Fare Distribution (Entire Population)

1. **Histogram by Gender**:
   - The histogram indicates that the majority of passengers, irrespective of gender, paid fares within the lower range, predominantly below 100.
   - Men show the highest frequency in fare distribution, with a significant spike at the lowest fare range (0-20).
   - Women have a lower frequency compared to men, with the highest concentration also in the lower fare range.
   - Children represent the smallest group, with a notable spike at the lowest fare range, similar to the other groups.
   - Overall, the fare distribution for all groups diminishes as the fare increases, with very few passengers paying fares above 200.

2. **Histogram by Class**:
   - The histogram reveals a clear distinction in fare distribution across different classes.
   - Third Class passengers, represented in green, show the highest frequency at the lowest fare range, with a prominent spike at 0-20.
   - Second Class passengers have a lower frequency compared to Third Class, with the distribution concentrated in the lower fare ranges.
   - First Class passengers, represented in blue, show a more spread-out distribution with fares extending up to 500, indicating a wider range of fare prices paid.
   - The distribution for First Class passengers is more even compared to the other classes, though the frequency still decreases as fare increases.
