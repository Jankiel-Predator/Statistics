# Examining Normality

### Objective
The goal of the exercise is to assess the normality of the sample distribution using a normal probability plot (QQ- plot).

## 1. Normal probability plot
A **normal probability plot**, also known as a Q-Q (quantile-quantile) plot, is a graphical technique used to assess whether a dataset follows a normal distribution. In this plot, the data's quantiles are plotted against the corresponding quantiles of a standard normal distribution. If the points form an approximately straight line, it indicates that the data is normally distributed. Deviations from this line suggest departures from normality.

## 2. Is the data normal?
1. **Normal Probability Plot Analysis**:
   - In the plot, the ordered values of the data are plotted against the theoretical quantiles of a standard normal distribution.
   - If the data is normally distributed, the points should lie approximately along the red straight line.
   - From the provided plot, the points appear to closely follow the straight line, suggesting that the data may be normally distributed.

2. **Comparison of Quantiles**:
   - The observed quantiles are compared with the theoretical quantiles. The closer these values are, the more likely it is that the data follows a normal distribution.
   - The observed and theoretical quantiles in the output show close alignment.

3. **Correlation Coefficient**:
   - The correlation coefficient between the data and the theoretical quantiles is 0.985, which is very close to 1.
   - A correlation coefficient close to 1 indicates a strong linear relationship, supporting the hypothesis that the data follows a normal distribution.

**Based on the normal probability plot and the high correlation coefficient (0.985), it is reasonable to conclude that the data appears to follow a normal distribution.**

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import probplot

plt.style.use("fivethirtyeight")
```

```python
data = [102, 105, 113, 120, 125, 127, 134, 135, 139, 141, 144, 145, 149, 150, 160]
```

```python
### Normal Probability Plot

plt.figure(figsize=(16,8))
probplot_data = probplot(data, plot=plt)
plt.title('Normal Probability Plot')
plt.show()
```

```python
### Comparison of Quantiles

print("Quantiles for the data:\n")
for i, (observed, theoretical) in enumerate(zip(probplot_data[0][1], probplot_data[0][0])):
    print(f"Observed Quantile {i + 1}: {observed:.2f} \t Theoretical Quantile {i + 1}: {theoretical:.2f}")
```

```python
### Correlation between quantiles and the data

quantiles = probplot_data[0][0]
correlation = np.corrcoef(data, quantiles)[0, 1]

print("Correlation coefficient for the data and standard normal distribution quantiles:", correlation)
```

   - If the data is normally distributed, the points should lie approximately along the red straight line.

2. **Comparison of Quantiles**:

   - The observed and theoretical quantiles in the output show close alignment.

3. **Correlation Coefficient**:
