## 1. What do those columns involve?
The Titanic dataset contains various columns, each representing specific details about the passengers:

- **survived**: Indicates whether the passenger survived (1) or did not survive (0).
- **pclass**: The passenger's class, with 1 being first class, 2 being second class, and 3 being third class.
- **sex**: The gender of the passenger (male or female).
- **age**: The age of the passenger in years.
- **sibsp**: The number of siblings or spouses the passenger had aboard the Titanic.
- **parch**: The number of parents or children the passenger had aboard the Titanic.
- **fare**: The fare paid for the ticket by the passenger.
- **embarked**: The port where the passenger boarded the Titanic (C = Cherbourg, Q = Queenstown, S = Southampton).
- **class**: The socio-economic status of the passenger, similar to the pclass column.
- **who**: Classification of the passenger (man, woman, child).
- **adult_male**: Indicates whether the passenger is an adult male (True) or not (False).
- **deck**: The deck on which the passenger's cabin was located.
- **embark_town**: The name of the town where the passenger boarded the Titanic.
- **alive**: Whether the passenger survived or not, similar to the survived column (yes or no).
- **alone**: Indicates if the passenger was traveling alone (True) or not (False).

## 2. How to retrieve information about the structure of a dataset?
In order to obtain the information about the number of entries in the dataset, the number of non-null values in each column and their data types, one may use `info()` method directly on the dataframe. From that, the columns may be divided into groups of data types:
- **Int** (integer values):
    - survived
    - pclass
    - sibsp
    - parch
- **Float** (numerical values with decimal points):
    - age
    - fare
- **Object** (text or combination of text and numbers in this case):
    - sex
    - emabrked
    - who
    - embark_town
    - alive
- **Category** (limited set of categories):
    - class
    - deck
- **Bool** (True or False):
    - adult_male
    - alone

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("fivethirtyeight")
```

```python
### Load Titanic dataset

titanic = sns.load_dataset('titanic')
```

```python
### Information about the structure of the dataframe

titanic.info()
```

```python
### Random sample of 10 rows from the dataframe

titanic.sample(10)
```

```python
### Filter rows with passengers who have relatives on board (sibsp > 0) or (parch > 0)

passengers_with_relatives = titanic[(titanic['sibsp'] > 0) | (titanic['parch'] > 0)]
passengers_with_relatives
```

# Histograms

```python
### Comparison of histograms of age for the subset of passengers and all

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))

# Plot histogram for passengers with relatives on board
sns.histplot(passengers_with_relatives['age'].dropna(), bins=20, color='skyblue', kde=True, ax=axes[0])
axes[0].set_title('Age of passengers with relatives on board')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

# Plot histogram for all passengers
sns.histplot(titanic['age'].dropna(), bins=20, color='salmon', kde=True, ax=axes[1])
axes[1].set_title('Age of all passengers')
axes[1].set_xlabel('Age')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

# Bar Plots

```python
### Survival bar plots for selected passengers and all

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot histogram for survival of passengers with relatives on board
sns.countplot(x='survived', data=passengers_with_relatives, hue='survived', 
              palette=['salmon', 'skyblue'], ax=axes[0], legend=False)
axes[0].set_title('Survival of passengers with relatives on board')
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['Did Not Survive', 'Survived'])
axes[0].set_xlabel('Survival')
axes[0].set_ylabel('Count')

# Plot histogram for survival of all passengers
sns.countplot(x='survived', data=titanic, hue='survived', 
              palette=['salmon', 'skyblue'], ax=axes[1], legend=False)
axes[1].set_title('Survival of all passengers')
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['Did Not Survive', 'Survived'])
axes[1].set_xlabel('Survival')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

# Dotplots

```python
### Dotplots of ticket prices for males and females with relatives onboard

# Divide the dataframe into males and females
males = passengers_with_relatives[passengers_with_relatives['sex'] == 'male']
females = passengers_with_relatives[passengers_with_relatives['sex'] == 'female']

# Using stripplot to create dotplots for males and females
plt.figure(figsize=(12, 6))
sns.stripplot(data=males, x='sex', y='fare', label='Males', color='skyblue')
sns.stripplot(data=females, x='sex', y='fare', label='Females', color='salmon')
plt.title("Dotplot of Ticket Prices for Males and Females with Relatives Onboard")
plt.xlabel("Sex")
plt.ylabel("Ticket Price (Fare)")
plt.legend()
plt.tight_layout()
plt.show()
```
