# Hypothesing Testing - Tests on Normality

### Objective
The objective is to assess the normality of the sample data. This will be achieved by visualizing the
data using Q-Q plot and Empirical Density Estimation. Subsequently, built-in Python tests including
Chi-square, Kolmogorov-Smirnov, Shapiro-Wilk, and Jarque-Bera will be employed to verify the
normality of the data.

## 1. Procedure
1. Visualization:
    - Generate a **Q-Q plot** to visually assess the normality of the sample data.
    - Construct an **Empirical Density Estimation plot** to further examine the distribution of the data.
2. Statistical Tests:
    - **Chi-square test**
    - **Kolmogorov-Smirnov test**
    - **Shapiro-Wilk test**
    - **Jarque-Bera test**

## 2. Formulation of Hypotheses

**Null Hypothesis (H₀):** The sample data follows a normal distribution.

**Alternative Hypothesis (H₁):** The sample data does not follow a normal distribution.

## 4. Conclusions

### 3.1. Sample Size Analysis

#### Small Sample Size (n=30)

- **Chi-square Test:** The p-values for both normal and uniform distributions are high, indicating no significant departure from normality.
- **Kolmogorov-Smirnov Test:** The p-values are relatively high, implying that the null hypothesis cannot be rejected, suggesting the sample data likely follows a normal distribution.
- **Shapiro-Wilk Test:** The p-values are consistently high, indicating no significant deviation from normality.
- **Jarque-Bera Test:** The p-values are high, suggesting that the null hypothesis cannot be rejected, indicating that the sample data is likely normally distributed.

#### Large Sample Size (n=1000)

- **Chi-square Test:** The p-values for both normal and uniform distributions are high, indicating no significant departure from normality.
- **Kolmogorov-Smirnov Test:** The p-values are relatively high for the normal distribution but extremely low for the uniform distribution, suggesting departure from normality in the latter case.
- **Shapiro-Wilk Test:** The p-values are consistently high, indicating no significant deviation from normality.
- **Jarque-Bera Test:** The p-values are high, suggesting that the null hypothesis cannot be rejected, indicating that the sample data is likely normally distributed.

### 3.2. Distribution Analysis

#### Normal Distribution

- Across both sample sizes, the p-values from all tests are generally high, indicating no significant departure from normality.
- The Jarque-Bera test provides the highest p-values, supporting the hypothesis that the data follows a normal distribution.

#### Uniform Distribution

- For small sample sizes, the p-values from most tests are high, indicating no significant departure from normality.
- However, for the large sample size (n=1000), the p-values from the Kolmogorov-Smirnov and Jarque-Bera tests are extremely low, suggesting departure from normality.
- The Chi-square test also indicates significant differences from normality for the large sample size, as evidenced by the extremely low p-value.

### 3.3. Real Data Analysis

- The p-values from all tests for the real data fall within a reasonable range.
- The Kolmogorov-Smirnov test indicates a departure from normality, as the p-value is very low.
- However, other tests such as Chi-square, Shapiro-Wilk, and Jarque-Bera provide relatively higher p-values, suggesting that the data might still be reasonably approximated by a normal distribution, albeit with some deviations.

## 5. Brief Overall Conclusions

- The tests on normality provide mixed results, indicating that the sample data's adherence to a normal distribution depends on both sample size and the distribution itself.
- For normal distributions, the sample data generally conforms to normality, especially for larger sample sizes.
- However, for uniform distributions, especially with large sample sizes, deviations from normality are observed, particularly in tests like Kolmogorov-Smirnov and Jarque-Bera.
- It's crucial to consider the context of the data and the assumptions of each test when interpreting these results.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

# Set a random seed for reproducibility
np.random.seed(42)

# Table of p-values
table = pd.DataFrame({"pvalue" : [0.9502, 0.3859, 0.7718, 0.5159,
                                        0.9057, 0.5679, 0.4772, 0.7148,
                                        0.0834, 0.8021, 0.7327, 0.3858,
                                        0.3056, 0.1298, 0.3189, 0.1574,
                                        0.9634, 0.0284, 0.2220, 0.7318]}, index=np.arange(1, 21))

# Function to conduct statistical tests
def test_normality(data):
    chi2_stat, chi2_p = stats.normaltest(data)
    ks_stat, ks_p = stats.kstest(data, 'norm')
    sw_stat, sw_p = stats.shapiro(data)
    jb_stat, jb_p = stats.jarque_bera(data)
    
    # Create DataFrame for results
    results = pd.DataFrame({
        'Statistic': [chi2_stat, ks_stat, sw_stat, jb_stat],
        'p-value': [chi2_p, ks_p, sw_p, jb_p]},
        index=['Chi-square', 'Kolmogorov-Smirnov', 'Shapiro-Wilk', 'Jarque-Bera'])
    return results

# Generate sample data from a normal distribution
normal_data_small = np.random.normal(loc=0, scale=1, size=30)
normal_data_large = np.random.normal(loc=0, scale=1, size=1000)

# Generate sample data from a uniform distribution
uniform_data_small = np.random.uniform(low=-1, high=1, size=30)
uniform_data_large = np.random.uniform(low=-1, high=1, size=1000)

# Visualization and Statistical Tests for Normal Data (Small Sample)
plt.figure(figsize=(12, 6))

# Q-Q Plot
plt.subplot(1, 2, 1)
stats.probplot(normal_data_small, dist="norm", plot=plt)
plt.title("Q-Q Plot - Normal Distribution (n=30)")

# Empirical Density Estimation Plot
plt.subplot(1, 2, 2)
plt.hist(normal_data_small, bins=10, density=True, alpha=0.6, color='g')
plt.title("Empirical Density Estimation - Normal Distribution (n=30)")

plt.tight_layout()
plt.show()

# Statistical Tests for Normal Data (Small Sample)
print("Normal Distribution (n=30):")
print(test_normality(normal_data_small))

# Visualization and Statistical Tests for Normal Data (Large Sample)
plt.figure(figsize=(12, 6))

# Q-Q Plot
plt.subplot(1, 2, 1)
stats.probplot(normal_data_large, dist="norm", plot=plt)
plt.title("Q-Q Plot - Normal Distribution (n=1000)")

# Empirical Density Estimation Plot
plt.subplot(1, 2, 2)
plt.hist(normal_data_large, bins=10, density=True, alpha=0.6, color='g')
plt.title("Empirical Density Estimation - Normal Distribution (n=1000)")

plt.tight_layout()
plt.show()

# Statistical Tests for Normal Data (Large Sample)
print("Normal Distribution (n=1000):")
print(test_normality(normal_data_large))

# Visualization and Statistical Tests for Uniform Data (Small Sample)
plt.figure(figsize=(12, 6))

# Q-Q Plot
plt.subplot(1, 2, 1)
stats.probplot(uniform_data_small, dist="norm", plot=plt)
plt.title("Q-Q Plot - Uniform Distribution (n=30)")

# Empirical Density Estimation Plot
plt.subplot(1, 2, 2)
plt.hist(uniform_data_small, bins=7, density=True, alpha=0.6, color='g')
plt.title("Empirical Density Estimation - Uniform Distribution (n=30)")

plt.tight_layout()
plt.show()

# Statistical Tests for Uniform Data (Small Sample)
print("Uniform Distribution (n=30):")
print(test_normality(uniform_data_small))

# Visualization and Statistical Tests for Uniform Data (Large Sample)
plt.figure(figsize=(12, 6))

# Q-Q Plot
plt.subplot(1, 2, 1)
stats.probplot(uniform_data_large, dist="norm", plot=plt)
plt.title("Q-Q Plot - Uniform Distribution (n=1000)")

# Empirical Density Estimation Plot
plt.subplot(1, 2, 2)
plt.hist(uniform_data_large, bins=10, density=True, alpha=0.6, color='g')
plt.title("Empirical Density Estimation - Uniform Distribution (n=1000)")

plt.tight_layout()
plt.show()

# Statistical Tests for Uniform Data (Large Sample)
print("Uniform Distribution (n=1000):")
print(test_normality(uniform_data_large))

# Visualization and Statistical Tests for Table Data
plt.figure(figsize=(12, 6))

# Q-Q Plot
plt.subplot(1, 2, 1)
stats.probplot(table["pvalue"], dist="norm", plot=plt)
plt.title("Q-Q Plot - Real Data")

# Empirical Density Estimation Plot
plt.subplot(1, 2, 2)
plt.hist(table["pvalue"], bins=10, density=True, alpha=0.6, color='g')
plt.title("Empirical Density Estimation - Real Data")

plt.tight_layout()
plt.show()

# Statistical Tests for Normal Data (Small Sample)
print("Real Data:")
print(test_normality(table["pvalue"]))
```
